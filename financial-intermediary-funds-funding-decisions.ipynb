```python
import pandas as pd
```

```python
df=pd.read_csv('financial-intermediary-funds-funding-decisions.csv')
```

```python
df
```

```python
df['As of Date'].unique()
```

```python
str_to_qtr = {
  'JAN-MAR': '3/31',
  'APR-JUN': '06/30',
  'JUL-SEP': '09/30',
  'OCT-DEC': '12/31'
}
```

```python
df['Approval Quarter'].unique()
```

```python
df['ApprovalQuarterCleaned']= df['Approval Quarter'].map(lambda x: str_to_qtr[x])
```

```python
df
```

```python
df.dtypes
```

```python
df['Calendar Year'].astype(str)
```

```python
df['ApprovalDateCleaned']=df['ApprovalQuarterCleaned'] + '/' + df['Calendar Year'].astype(str)
```

```python
df
```

```python
df['ApprovalDateCleaneddt'] =  pd.to_datetime(df['ApprovalDateCleaned'], format='%m/%d/%Y')
```

```python
df.dtypes
```

```python
df
```

```python
columns = ['As of Date', 'Approval Quarter', 'Calendar Year','ApprovalQuarterCleaned','ApprovalDateCleaned']
df.drop(columns, inplace=True, axis=1)
```
